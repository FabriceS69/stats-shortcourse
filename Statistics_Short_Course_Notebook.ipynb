### Remember this:
How many full houses in a deck of cards?

```python
# make the deck
suits = ['hearts', 'diamonds', 'spades', 'clubs']
values = ['2', '3', '4', '5', '6', '7', '8', '9',
          '10', 'Jack', 'Queen', 'King', 'Ace']
deck = []
for suit in suits:
    for value in values:
        deck.append(value + " of " + suit)
```

```python
deck
```

```python
hands = list(combinations(deck, 5))
print(len(hands))
```

```python
def make_full_house_list(hands):
    full_houses = []
    for hand in hands:
        values = [val.split(' ')[0] for val in hand]
        num_values = set(values)
        if len(num_values) == 2:
            # check if full house for 4 of a kind
            first_val = values[0]
            num_1stval = values.count(first_val)
            if (num_1stval == 2 or num_1stval == 3):
                full_houses.append(hand)
    return full_houses
```

```python
full_houses = make_full_house_list(hands)
```

```python
full_houses
```

```python
len(full_houses)
```

### Probability Distributions

```python
# make some randomly distributed grades
sigma = 10
mu = 75
x = sigma* np.random.randn(200) + mu

counts, _ = np.histogram(x, bins=25, range=(50,100))
vals = np.arange(51,101,2)

# plot it
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(vals, counts)
ax.set_xlim(50, 100)
ax.set_xlabel('x')
ax.set_ylabel('grade count');
```

```python
# A student from a different class gets a score of 100 on the test.
# How likely is a score like that from our class?
# If all we used was the histogram above we would simply say it's impossible.
#
# Let's make a distribution and see how well it fits the data
x_bar = np.mean(x)  # average value of our data
s = np.std(x,ddof=1) # standard deviation of our data
```

```python
from scipy import stats
```

```python
dist = stats.norm(loc=x_bar, scale=s) # trying normal distribution first
```

```python
x_pdf = np.linspace(dist.ppf(0.001), dist.ppf(0.999), 200)  # seeing 99.8% of the area
pdf = dist.pdf(x_pdf)
```

```python
# scaling
max_count = max(counts)
max_pdf = max(pdf)

# plot them both
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(vals, counts/max_count, label='Class scores')
ax.plot(x_pdf, pdf/max_pdf, 'k-', label = 'Assumed distribution')
ax.set_xlim(50, 100)
ax.set_xlabel('x')
ax.set_ylabel('grade count')
ax.set_title('Comparison of class scores with normal distribution')
ax.legend();
```

### It's very unlikely that a student could have scored 100 from our class.

```python
area_to_100 = dist.cdf(100)
area_to_100
```

```python
chance_percent = (1 - area_to_100) * 100
# chance of 100 or greater
print("The chance of a perfect score from these students "
      f"was {chance_percent:0.5f}%.")
```
